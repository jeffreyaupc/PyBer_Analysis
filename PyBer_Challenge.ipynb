# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"].sum()
print(f"Urban total rides: {urban_ride_count}")

suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"].sum()
print(f"Suburban total rides: {suburban_ride_count}")

rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"].sum()
print(f"Rural total rides: {rural_ride_count}")
```

```python
# 2. Get the total drivers for each city type
urban_total_drivers = urban_cities_df.groupby(["city"]).mean()["driver_count"].sum()
print(f"Urban total drivers: {urban_total_drivers}")

suburban_total_drivers = suburban_cities_df.groupby(["city"]).mean()["driver_count"].sum()
print(f"Suburban total drivers: {suburban_total_drivers}")

rural_total_drivers = rural_cities_df.groupby(["city"]).mean()["driver_count"].sum()
print(f"Rural total drivers: {rural_total_drivers}")
```

```python
#  3. Get the total amount of fares for each city type
urban_fare = urban_cities_df.groupby(["city"])["fare"].sum().sum()
print(f"Urban fare: {urban_fare}")

suburban_fare = suburban_cities_df.groupby(["city"])["fare"].sum().sum()
print(f"Suburban fare: {suburban_fare}")

rural_fare = rural_cities_df.groupby(["city"])["fare"].sum().sum()
print(f"Rural fare: {rural_fare}")
```

```python
#  4. Get the average fare per ride for each city type. 
urb_avg_fare_per_ride = urban_fare / urban_ride_count
print(f"Urban average fare per ride: {urb_avg_fare_per_ride}")

suburb_avg_fare_per_ride = suburban_fare / suburban_ride_count
print(f"Suburban average fare per ride: {suburb_avg_fare_per_ride}")

rural_avg_fare_per_ride = rural_fare / rural_ride_count
print(f"Rural average fare per ride: {rural_avg_fare_per_ride}")
```

```python
# 5. Get the average fare per driver for each city type. 
urb_avg_fare_per_driver = urban_fare / urban_total_drivers
print(f"Urban average fare per driver: {urb_avg_fare_per_driver}")

suburb_avg_fare_per_driver = suburban_fare / suburban_total_drivers
print(f"Suburban average fare per driver: {suburb_avg_fare_per_driver}")

rural_avg_fare_per_driver = rural_fare / rural_total_drivers
print(f"Rural average fare per driver: {rural_avg_fare_per_driver}")
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    "type": ["Rural", "Suburban", "Urban"],
    "Total Rides": [rural_ride_count, suburban_ride_count, urban_ride_count],
    "Total Drivers": [rural_total_drivers, suburban_total_drivers, urban_total_drivers],
    "Total Fares": [rural_fare, suburban_fare, urban_fare],
    "Average Fare per Ride": [rural_avg_fare_per_ride, suburb_avg_fare_per_ride, urb_avg_fare_per_ride],
    "Average Fare per Driver": [rural_avg_fare_per_driver, suburb_avg_fare_per_driver, urb_avg_fare_per_driver]
}).set_index("type")
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,.0f}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_summary_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_data_df = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
pyber_data_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
pyber_data_df = pyber_data_df.reset_index()
pyber_data_df.head()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_data_df_pivot = pyber_data_df.pivot(index="date", columns="type", values="fare")
pyber_data_df_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_pivot_jan_apr = pyber_data_df_pivot.loc['2019-01-01':'2019-04-28']
pyber_pivot_jan_apr
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
pyber_pivot_jan_apr.index = pd.to_datetime(pyber_pivot_jan_apr.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_pivot_jan_apr.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_df = pyber_pivot_jan_apr.resample('W').sum()
weekly_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
weekly_fares = weekly_fares_df.plot(figsize = (20, 6))
weekly_fares.set_title('Total Fare by City Type',fontsize=20)
weekly_fares.set_ylabel('Fare ($USD)',fontsize=14)
weekly_fares.set_xlabel("",fontsize=14)
weekly_fares.grid(True)


# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()
```
